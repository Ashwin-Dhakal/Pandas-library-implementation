```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python

df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df= pd.merge(df1, df2, on = "city")
df
```

```python
dfinner= pd.merge(df1, df2, on = "city", how = "left", indicator=True) #indicator le kata bata data aako bhanera bhancha  
dfinner
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_ASHWIN','_ALOK')) #it creates the first second value for the line segment
df3
```
