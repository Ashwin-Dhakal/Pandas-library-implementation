```python
import pandas as pd
df = pd.read_csv("weather_data.csv")
df
#here it contains no data values so we need to haldle the missing datas 
```

```python
# option 1: fill all the missing values with zeros
new_df = df.fillna(0)
new_df
```

```python
# option 2: replace with different colums values
second_df = df.fillna({
    'temperature' : 25,
    'windspeed' : 5,
    'event' : "no idea",
})
second_df
```

```python
#option 3: replace the null value with the pervious day value
third_df = df.fillna(method = "ffill") #ffill - forward fill
third_df
```

```python
#option 4: replace the null value with the next day value
forth_df = df.fillna(method = "bfill") #bfill - backward fill
forth_df
```

```python
#option 5: replace the null value with the intetrpolation value 
fifth_df = df.interpolate()
fifth_df
```

```python
#option 6: droping all the non values
#better use threshold and how parameters
sixth_df = df.dropna()
sixth_df
```
